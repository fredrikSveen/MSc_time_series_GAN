# Complex sine generation

```python
# Necessary imports
import numpy as np
import pandas as pd
import json
```

### The original complex sines

```python
comp_sine_6_1000 = np.loadtxt(f'complex_sine_data/comp_sine_123_1000_6.csv', delimiter = ",",skiprows = 1) # Complex sine
df_comp_sine_6_1000 = pd.DataFrame(comp_sine_6_1000)
df_comp_sine_6_1000.plot(subplots=True)
df_comp_sine_6_1000.iloc[:100].plot(subplots=True) # Zoom in on the first 200 samples
```

### TimeGAN
The synthetic complex sines generated by TimeGAN

```python
with open('synthetic_complex_sines/timegan/comp_syn_sine_1000_6_100_110524_14h20.json', 'r') as file:
    syn_comp_1000_6 = json.load(file)
with open('synthetic_complex_sines/timegan/comp_syn_sine_10000_6_100_110524_17h26.json', 'r') as file:
    syn_comp_10000_6 = json.load(file)
with open('synthetic_complex_sines/timegan/comp_syn_sine_50000_6_100_120524_08h13.json', 'r') as file:
    syn_comp_50000_6 = json.load(file)
with open('synthetic_complex_sines/timegan/comp_syn_sine_100000_6_100_140524_12h06.json', 'r') as file:
    syn_comp_100k_6 = json.load(file)

df_syn_comp_1000_6 = pd.DataFrame(syn_comp_1000_6[0])
df_syn_comp_10000_6 = pd.DataFrame(syn_comp_10000_6[0])
df_syn_comp_50000_6 = pd.DataFrame(syn_comp_50000_6[0])
df_syn_comp_100k_6 = pd.DataFrame(syn_comp_100k_6[0])
```

```python
df_syn_comp_1000_6.plot(subplots=True, title="1000 iterations")
df_syn_comp_10000_6.plot(subplots=True, title="10 000 iterations")
df_syn_comp_50000_6.plot(subplots=True, title="50 000 iterations")
df_syn_comp_100k_6.plot(subplots=True, title="100 000 iterations")
```
