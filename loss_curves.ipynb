```python
file1 = open('loss_evolution.txt', 'r')

losses = {' d_loss': [], ' g_loss_u' : [], ' g_loss_s': [], ' g_loss_v': [], ' e_loss_t0': []}
for line in file1:
    tmp1 = line.strip().split(',')
    tmp2 = [i.split(':') for i in tmp1]
    for pair in tmp2:
        if pair[0] == 'step':
            continue
        losses[pair[0]].append(float(pair[1].strip()))
file1.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 16})

labels = ['Discriminator loss', ' g_loss_u', ' g_loss_s', ' g_loss_v', ' e_loss_t0']

x = np.arange(0, 100000, 1000)
cnt = 0
for key, value in losses.items():
    plt.figure(figsize=(12,6))
    plt.plot(x, value[1:101])
    plt.xlabel('Iterations')
    plt.ylabel(labels[cnt])
    cnt += 1
    plt.show()
```
