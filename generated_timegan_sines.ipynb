```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import MinMaxScaler
```

### The original sines used as input

```python
def plot_loc_right(df, title=None, subplots=False, figsize=(12,8)):
    axes = df.plot(title=title, subplots=subplots, figsize=figsize)
    axes = axes.flat  # .ravel() and .flatten() also work

    # extract the figure object to use figure level methods
    fig = axes[0].get_figure()

    # iterate through each axes to use axes level methods
    for ax in axes:
        
        ax.legend(loc='right', fontsize=14)
        
    plt.show()

dim = 9
plt.rcParams.update({'font.size': 14})
ori_data = np.loadtxt(os.path.join('sine_data', f'sine_123_1000_{dim}.csv'), delimiter = ",",skiprows = 1)
clip_length = 200

scaler = MinMaxScaler()
ori_data = pd.DataFrame(scaler.fit_transform(ori_data))


axes = ori_data[:clip_length].plot(subplots=True, figsize=(12,8))
axes = axes.flat  # .ravel() and .flatten() also work

# extract the figure object to use figure level methods
fig = axes[0].get_figure()

# iterate through each axes to use axes level methods
for ax in axes:
    
    ax.legend(loc='right', fontsize=14)
    
plt.show()
```

### Resulting single sines for different number of iterations with sequence length 25

```python
with open('synt_data/timeGan_generated_sines/syn_sine_1000_1_25_020524_10h43.json', 'r') as file:
    python_obj1 = json.load(file)
with open('synt_data/timeGan_generated_sines/syn_sine_10000_1_25_230424_11h35.json', 'r') as file:
    python_obj2 = json.load(file)
with open('synt_data/timeGan_generated_sines/syn_sine_50000_1_25_230424_15h22.json', 'r') as file:
    python_obj3 = json.load(file)
```

```python
df1 = pd.DataFrame(python_obj1[0])
df2 = pd.DataFrame(python_obj2[0])
df3 = pd.DataFrame(python_obj3[0])

df1[1].plot(figsize=(12,6))
df2[1].plot()
df3[1].plot()
ori_data[:25][0].plot()
plt.legend(["1000 iterations","10 000 iterations", "50 000 iterations", "Original sine"])
plt.title(f'25 samples long 1D sine generated by TimeGAN with different number of iterations')
plt.show()
```

Discussion: Struggles to get the correct form with only 1000 iterations, but for 10 000 and 50 000 it generates very nice sine waves.
