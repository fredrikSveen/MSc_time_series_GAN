```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

### Resulting single sines for different number of iterations with sequence length 25

```python
with open('synt_data/timeGan_generated_sines/syn_sine_1000_1_230424_11h01.json', 'r') as file:
    python_obj1 = json.load(file)
with open('synt_data/timeGan_generated_sines/syn_sine_10000_1_230424_11h35.json', 'r') as file:
    python_obj2 = json.load(file)
with open('synt_data/timeGan_generated_sines/syn_sine_50000_1_230424_15h22.json', 'r') as file:
    python_obj3 = json.load(file)
```

```python
df1 = pd.DataFrame(python_obj1[0])
df2 = pd.DataFrame(python_obj2[0])
df3 = pd.DataFrame(python_obj3[0])

df1[1].plot()
df2[1].plot()
df3[1].plot()
plt.legend(["1000 iterations","10 000 iterations", "50 000 iterations"])
plt.show()
```

Discussion: Struggles to get the correct form with only 1000 iterations, but for 10 000 and 50 000 it generates very nice sine waves.
